# Astroid Impact

## This project analyzes near-Earth asteroids using data from NASA's API.

### This project was inspired by my curiosity about asteroid impacts, close approaches to Earth, and the potential risks these objects may pose to our planet and the satellites that support modern communication, navigation, and scientific research.

```python
import prettytable
import pandas as pd
import sqlite3 as sql3
import requests
import plotly.express as px
import datetime as dt
```

```python
prettytable.DEFAULT = 'DEFAULT'

conn = sql3.connect("AsteroidDB.db")
cursor = conn.cursor()

%load_ext sql

current = dt.datetime.now()
to_date = current + dt.timedelta(days=6)
current_str = dt.datetime.now().strftime("%Y-%m-%d")
to_str = to_date.strftime("%Y-%m-%d")
```

## DATA COLLECTION

### Using NASA's Near-Earth Object (NEO) API, I collected data on asteroids approaching Earth. The dataset includes the asteroid's name, minimum and maximum estimated diameter in feet, hazardous classification, close approach date and time, and miss distance in miles. This information was extracted from the API's JSON response and stored in a Pandas DataFrame for further analysis.


```python
api_key = 'YyIyPcf4iQORkZfTv7DFdaZw7KqSQ6KlSSeso2Zq'
url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date={current_str}&end_date={to_str}&api_key={api_key}"

response = requests.get(url)
data = response.json()
ob_data = data['near_earth_objects']
```

## DATA CLEANING

```python
asteroid_data = pd.DataFrame(columns=['Name', 'Min Diameter (feet)', 'Max Diameter (feet)', 'Hazardous', 'Close Approach Date', 'Miss Distance (miles)'])
```

```python
for item in ob_data:
    for i, ob in enumerate(ob_data[item]):
        name = ob_data[item][i]['name'].split("(")[1].rstrip(")")
        min_diameter = round(ob_data[item][i]['estimated_diameter']['feet']['estimated_diameter_min'],2)
        max_diameter = round(ob_data[item][i]['estimated_diameter']['feet']['estimated_diameter_max'],2)
        is_hazardous = ob_data[item][i]['is_potentially_hazardous_asteroid']
        close_approach_date = pd.to_datetime(ob_data[item][i]['close_approach_data'][0]['close_approach_date_full'])
        miss_distance = round(float(ob_data[item][i]['close_approach_data'][0]['miss_distance']['miles']),2)
        asteroid_data = pd.concat([asteroid_data,pd.DataFrame({'Name':[name],
                                                             'Min Diameter (feet)':[min_diameter], 
                                                             'Max Diameter (feet)':[max_diameter],
                                                             'Hazardous':[is_hazardous],
                                                             'Close Approach Date':[close_approach_date],
                                                             'Miss Distance (miles)':[miss_distance]})], ignore_index=True)
```

```python
asteroid_data.dropna(inplace=True)
asteroid_data
```

## DATABASE CREATION

### After cleaning and organizing the asteroid data in a Pandas DataFrame, the dataset was loaded into a SQLite database. Storing the data in a relational database enabled the performance of SQL queries, calculation of summary statistics, filtering of records, and identification of patterns within the dataset. The database serves as a structured and reusable source for analysis and visualization.

```python
asteroid_data.to_sql("ASTEROID_IMPACT_DATA", conn, if_exists='replace', index=False, method='multi');
```

```python
%sql sqlite:///AsteroidDB.db
```

```python
%sql select name, type, length(type) from PRAGMA_TABLE_INFO("ASTEROID_IMPACT_DATA");
```

## SQL ANALYSIS

### This query retrieves all asteroids expected to make a close approach to Earth within the next seven days.

```python
%sql select *, CASE when hazardous = 1 then 'True' else 'false' end as hazardous from ASTEROID_IMPACT_DATA;
```

### This query retrieves the name, close approach date, and miss distance of all asteroids classified as hazardous that are expected to make a close approach to Earth within the next seven days.

```python
%sql select name, date([Close Approach Date]) as "Approach Date", time([Close Approach Date]) \
as "Approach Time", "miss distance (miles)" from ASTEROID_IMPACT_DATA where hazardous = true group by name;
```

### This query retrieves the ten asteroids with the smallest miss distances, along with their close approach date and time and hazardous classification. The results help identify which asteroids are expected to pass closest to Earth during the observation period.

```python
%sql select name, printf('%.2f', "miss distance (miles)", 2) as "Miss Distance", date([Close Approach Date]) \
as "Approach Date", time([Close Approach Date]) as "Approach Time", case when hazardous = 1 \
then 'true' else 'false' end as hazardous from ASTEROID_IMPACT_DATA \
order by [miss distance (miles)] asc limit 10;
```

### This query retrieves the names of the ten largest asteroids expected to make a close approach to Earth within the next seven days, along with their estimated maximum diameter and hazardous classification. The results help identify the largest objects in the dataset and assess whether any are considered potentially hazardous.

```python
%sql select name, printf('%.2f', [Max Diameter (feet)]) as [Maximum Diameter by Feet], \
CASE when Hazardous = 1 THEN 'true' ELSE 'false' end as Hazardous from ASTEROID_IMPACT_DATA order by [Max Diameter (feet)] desc \
limit 10;  
```

### This query calculates the average miss distance of asteroids expected to make a close approach to Earth within the next seven days. The result provides a general measure of how close approaching asteroids are expected to come to Earth during the observation period.

```python
%sql select round(avg([Miss Distance (miles)]), 2) AS [Average Miss Distance by Feet] from ASTEROID_IMPACT_DATA;
```

### This query calculates the average estimated diameter of asteroids expected to make a close approach to Earth within the next seven days. The result provides an overall measure of the typical size of asteroids in the dataset during the observation period.

```python
%sql select round(avg([Max Diameter (feet)]), 2) as "Average Max Diameter (feet)", round(avg([Min Diameter (feet)]), 2) as "Average Min Diameter (feet)" from ASTEROID_IMPACT_DATA;
```

## Visualizations

```python
fig_asteroid_miss_distance = px.scatter(
    asteroid_data,
    x = 'Miss Distance (miles)',
    y = 'Max Diameter (feet)',
    color= 'Hazardous',
    title = 'Distribution of Asteroid Sizes'
)
fig_asteroid_miss_distance.show()
```

```python
fig_asteroid_sizes = px.histogram(
    asteroid_data,
    x= 'Max Diameter (feet)',
    title= 'Distribution of Asteroid Sizes'
)
fig_asteroid_sizes.show()
```

```python
haz_counts = asteroid_data['Hazardous'].value_counts().reset_index()
fig_asteroid_approach_date = px.bar(
    haz_counts,
    x = 'Hazardous',
    y = 'count',
    title = 'Asteroids Hazardous to the Earth'
)
fig_asteroid_approach_date.show()
```

## FINDINGS

### The analysis showed that most asteroids approaching Earth within the seven-day observation period were not classified as hazardous. While several large asteroids were identified, size alone did not determine whether an asteroid was considered hazardous. The closest asteroids in the dataset still maintained significant distances from Earth, indicating no immediate impact threat during the observation period.

### The average asteroid diameter and average miss distance provided additional context about the typical objects detected by NASA's Near-Earth Object monitoring program. By examining asteroid size, proximity, and hazardous classification together, it was possible to gain a better understanding of the characteristics of near-Earth objects currently being tracked.
